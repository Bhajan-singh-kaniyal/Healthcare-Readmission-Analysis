# 🏥 Healthcare Readmission Analysis

## Objective
To analyze patient data and identify factors affecting hospital readmission.

## Dataset
Diabetes 130-US hospitals dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/cleaned_data.csv")
df.head()
```

```python
df.info()
```

### Observation
- Dataset contains patient medical records
- Data includes both numerical and categorical columns

```python
sns.countplot(x='readmitted', data=df)
plt.title("Readmission Distribution")
plt.show()
```

### Insight
- Many patients are getting readmitted
- This indicates a major healthcare concern

```python
sns.histplot(df['age'])
plt.title("Age Distribution")
plt.show()
```

### Insight
- Most patients belong to older age groups
- Elderly patients have higher readmission risk

## 💡 Key Insights

- Elderly patients are more likely to be readmitted  
- Frequent hospital visits increase readmission chances  
- Patients with more medications are at higher risk  

## 📈 Conclusion

This analysis helps hospitals identify high-risk patients and improve care.
